```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import warnings

warnings.filterwarnings('ignore')
```

```python
#Lets read temperature file to create the DF
file_temp = "weather_data/temperature.csv"
temp_df = pd.read_csv(file_temp, encoding="ISO-8859-1")
temp_df_city = temp_df[['datetime','Los Angeles','Chicago','Vancouver','Philadelphia','New York']]
temp_df_city_2014 = temp_df_city[temp_df_city['datetime'].str.contains('2014')]
temp_df_city
#Remove the NAN. Since we have lot of data we can just drop any values
temp_df_city_2014 = temp_df_city_2014.dropna(how="any")
temp_df_city_2014.head()
```

```python
temp_df_city_2014[['Year','Month','Ext']] = temp_df_city_2014['datetime'].str.split('-', expand=True)
temp_df_city_2014[['Day','Time']] = temp_df_city_2014['Ext'].str.split(' ', expand=True)
temp_df_city_2014 = temp_df_city_2014.drop(['Ext'], axis=1)
temp_df_city_2014.head()
```

```python
temp_df_city_2014_month = temp_df_city_2014.groupby(['Month'])
temp_df_city_2014_month_avg = temp_df_city_2014_month.mean()
temp_df_city_2014_month_avg
```

```python
#Lets read crime rate for Los Angeles
crime_file = "crime-Los Angeles/LA_CRIME_2010_2017.csv"
crime_LA = pd.read_csv(crime_file, encoding="us-ascii")
crime_LA_2014 = crime_LA[crime_LA['Date Occurred'].str.contains('2014')]
crime_LA_2014
```

```python
crime_LA_2014[['Month','Day','Year']] = crime_LA_2014['Date Occurred'].str.split('/', expand=True)
crime_LA_2014.head()
```

```python
crime_LA_2014_month = crime_LA_2014.groupby('Month')
crime_LA_2014_month_count_df = crime_LA_2014_month.count()
crime_LA_2014_month_count_df
```

```python
temp_df_city_2014_month_avg['Crime Los Angeles'] = crime_LA_2014_month_count_df['Date Occurred']
temp_df_city_2014_month_avg
```

```python
#Lets read crime rate for Chicago
crime_file = "crime-Chicago/CHI_CRIME_2012_2017.csv"
crime_CHI = pd.read_csv(crime_file, encoding="us-ascii")

crime_CHI_2014 = crime_CHI[crime_CHI['Date'].str.contains('2014')]
crime_CHI_2014
```

```python
crime_CHI_2014[['Month','Day','Ext']] = crime_CHI_2014['Date'].str.split('/', expand=True)
crime_CHI_2014[['Year','Time']] = crime_CHI_2014['Ext'].str.split(' ', 1, expand=True)
crime_CHI_2014 = crime_CHI_2014.drop(['Ext'], axis=1)

crime_CHI_2014.head()
```

```python
crime_CHI_2014_month = crime_CHI_2014.groupby('Month')
crime_CHI_2014_month_count_df = crime_CHI_2014_month.count()
crime_CHI_2014_month_count_df
```

```python
temp_df_city_2014_month_avg['Crime Chicago'] = crime_CHI_2014_month_count_df['Date']
temp_df_city_2014_month_avg
```

```python
#Lets read crime rate for New York
crime_file = "crime-New York/NYPD_Complaint_Data_Historic.csv"
crime_NYC = pd.read_csv(crime_file, encoding="us-ascii")
crime_NYC_2014 = crime_NYC[crime_NYC['RPT_DT'].str.contains('2014')]
crime_NYC_2014.head()
```

```python
crime_NYC_2014[['Month','Day','Year']] = crime_NYC_2014['RPT_DT'].str.split('/', expand=True)
crime_NYC_2014.head()
```

```python
crime_NYC_2014['Month'] = crime_NYC_2014['Month'].astype(str).str.zfill(2)
crime_NYC_2014_month = crime_NYC_2014.groupby('Month')
crime_NYC_2014_month_count_df = crime_NYC_2014_month.count()
crime_NYC_2014_month_count_df
```

```python
temp_df_city_2014_month_avg['Crime NYC'] = crime_NYC_2014_month_count_df['RPT_DT']
temp_df_city_2014_month_avg
```

```python
#Lets read crime rate for Philadelphia
crime_file = "crime-Philadelphia/crime.csv"
crime_PHI = pd.read_csv(crime_file, encoding="us-ascii")

crime_PHI_2014 = crime_PHI[crime_PHI['Dispatch_Date'].str.contains('2014')]
crime_PHI_2014
```

```python
crime_PHI_2014[['Year','Month','Date']] = crime_PHI_2014['Dispatch_Date'].str.split('-', expand=True)
crime_PHI_2014.head()
```

```python
crime_PHI_2014_month = crime_PHI_2014.groupby('Month')
crime_PHI_2014_month_count_df = crime_PHI_2014_month.count()
crime_PHI_2014_month_count_df
```

```python
temp_df_city_2014_month_avg['Crime Philadelphia'] = crime_PHI_2014_month_count_df['Dispatch_Date']
temp_df_city_2014_month_avg
```

```python
#Lets read crime rate for Vancouver
crime_file = "crime-Vancouver/crime.csv"
crime_VAN = pd.read_csv(crime_file, encoding="us-ascii")
crime_VAN.head()

crime_VAN_2014 = crime_VAN.loc[crime_VAN['YEAR'] == 2014]
crime_VAN_2014
```

```python
#df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
crime_VAN_2014 = crime_VAN_2014.rename(columns={'YEAR':'Year',
                                                'MONTH':'Month',
                                               'DAY': 'Day'})
crime_VAN_2014.head()
```

```python
crime_VAN_2014['Month'] = crime_VAN_2014['Month'].astype(str).str.zfill(2)
crime_VAN_2014_month = crime_VAN_2014.groupby('Month')
crime_VAN_2014_month_count_df = crime_VAN_2014_month.count()
crime_VAN_2014_month_count_df
```

```python
temp_df_city_2014_month_avg['Crime Vancouver'] = crime_VAN_2014_month_count_df['TYPE']
temp_df_city_2014_month_avg
```

```python
# Write to Output File (CSV)
output_data_file = f'output_data/TempVsCrime-2014.csv'
temp_df_city_2014_month_avg.to_csv(output_data_file, index=False)
```
